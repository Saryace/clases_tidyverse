## Ejemplo wide: graficar usando la función `ggplot2`

```r
penguins %>% # gráfico de masas de las hembras en dos islas: Biscoe y Dream
  filter(island %in% c("Biscoe", "Dream") & sex == "female") %>%
  ggplot(aes(x = species, y = flipper_length_mm)) + #solo hembras y solo flipper_length_mm
  geom_boxplot(aes(color = species), width = 0.3, show.legend = FALSE) +
  scale_color_manual(values = c("darkorange","purple","cyan4")) +
  theme_minimal() +
  labs(x = "Especies",
       y = "Peso (g)")
```

```r
penguins %>% #empecemos quitando la función filtro
  ggplot(aes(x = species, y = flipper_length_mm)) + # todos los pingüinos y flipper_length_mm
  geom_boxplot(aes(color = sex), width = 0.3, show.legend = FALSE) + # color ahora es el sexo
  scale_color_manual(values = c("darkorange","purple","cyan4")) +
  theme_minimal() +
  labs(x = "Especies",
       y = "Peso (g)")
```

## Ejemplo long: graficar usando la función `ggplot2` para varias columnas

```r
penguins %>% 
    pivot_longer(cols = contains("_"), # todas las mediciones tienen "_", asi las seleccionamos asi
                 names_to = c("parte_cuerpo", "medicion","unidad"), # información del título de la columna
                 values_to = "valores", # nueva columna
                 names_sep = "_") %>%  # separador que usaremos
    tail(n = 5)
```

![gif](https://github.com/gadenbuie/tidyexplain/raw/main/images/tidyr-pivoting.gif)

## Ahora podemos graficar las diferentes "mediciones"

```r
penguins %>% 
    pivot_longer(cols = contains("_"), 
                 names_to = c("parte_cuerpo", "medicion","unidad"), 
                 values_to = "valores", 
                 names_sep = "_") %>%  
ggplot(aes(x = sex, y = valores)) +
    geom_boxplot(aes(color = island), width = 0.3, show.legend = FALSE) + # color ahora es el sexo
    scale_color_manual(values = c("darkorange","purple","cyan4")) +
    facet_wrap(.~ medicion + parte_cuerpo, scales = "free") +
    theme_minimal()
        
```
